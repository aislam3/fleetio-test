```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
leads = pd.read_csv("leads.csv")
```

```python
leads = pd.DataFrame(leads)
leads.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier

from IPython.display import display
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
```

```python
leads = leads.assign(converted = leads['converted_at'].notnull().astype("int"))

```

```python
leads.head()
```

```python
#Define training and validation split
dep_var = "converted" #sets target
cond = np.random.rand(len(leads)) > .2 #sets training to ~80% of data
train = np.where(cond)[0] #(training at 80%)
valid = np.where(~cond)[0] #validation at 20%

len(train), len(valid) #check size
```

```python
train_df = leads.iloc[train] #get training dataframe
valid_df = leads.iloc[valid] #get validation dataframe
len(train_df), len(valid_df)
```
